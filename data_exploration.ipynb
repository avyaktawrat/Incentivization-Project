```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_textblob = np.loadtxt('.\GTwitter-20201020T083256Z-001\opinion_textblob.txt',skiprows=0)
data_tweetrain = np.loadtxt('.\GTwitter-20201020T083256Z-001\opinion_tweet_rain.txt',skiprows=0)
data_opinion = np.loadtxt('.\GTwitter-20201020T083256Z-001\opinion.txt',skiprows=0)
```

```python
nodeId_1, countId_1 = np.unique(data_textblob[:,0],return_counts=True)
nodeId_2, countId_2 = np.unique(data_tweetrain[:,0],return_counts=True)
nodeId_3, countId_3 = np.unique(data_opinion[:,0],return_counts=True)
print("Number of unique users in textblob, tweet rain and opinions are given by ",np.shape(nodeId_1),np.shape(nodeId_2),np.shape(nodeId_3)," respectively.")
```

```python
nodeId_1 = np.reshape(nodeId_1,[nodeId_1.shape[0],1])
countId_1 = np.reshape(countId_1,[countId_1.shape[0],1])
nodeId_2 = np.reshape(nodeId_2,[nodeId_2.shape[0],1])
countId_2 = np.reshape(countId_2,[countId_2.shape[0],1])
nodeId_3 = np.reshape(nodeId_3,[nodeId_3.shape[0],1])
countId_3 = np.reshape(countId_3,[countId_3.shape[0],1])

data_counts_1 = np.append(nodeId_1,countId_1, axis = 1)
data_counts_2 = np.append(nodeId_2,countId_2, axis = 1)
data_counts_3 = np.append(nodeId_3,countId_3, axis = 1)
print(np.shape(data_counts_1),np.shape(data_counts_2),np.shape(data_counts_3))
```

```python
user_count_1, post_count_1 = np.unique(data_counts_1[:,1],return_counts=True)
user_count_2, post_count_2 = np.unique(data_counts_2[:,1],return_counts=True)
user_count_3, post_count_3 = np.unique(data_counts_3[:,1],return_counts=True)

print("Number of points in #users vs #posts is ", np.shape(user_count_1)[0],",",np.shape(user_count_2)[0],",",np.shape(user_count_3)[0], " in textblob, tweet rain and opinions respectively.")
```

Common Plot for all 3 datasets

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(post_count_1, user_count_1, color='b',s=6,alpha=0.5)
ax.scatter(post_count_2, user_count_2, color='r',s=6,alpha=0.7)
ax.scatter(post_count_3, user_count_3, color='g',s=6,alpha=0.5)
ax.set_xlabel('Number of Posts')
ax.set_ylabel('Number of Users')
ax.set_title('Scatter Plot')
plt.show()
```

Separate Plots for all three datasets

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(20,5))

ax1.scatter(post_count_1, user_count_1, color='b',s=6,alpha=0.5)
ax2.scatter(post_count_2, user_count_2, color='r',s=6,alpha=0.7)
ax3.scatter(post_count_3, user_count_3, color='g',s=6,alpha=0.5)
ax1.set_title('Textblob')
ax2.set_title('Tweet Rain') 
ax3.set_title('Opinions')
```
