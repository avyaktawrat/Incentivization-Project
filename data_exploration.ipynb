```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('.\GTwitter-20201020T083256Z-001\opinion_textblob.txt',skiprows=0)
nodeId, countId = np.unique(data[:,0],return_counts=True)
print(np.shape(nodeId))
```

```python
data_counts = np.append(np.reshape(nodeId,[nodeId.shape[0],1]), np.reshape(countId,[countId.shape[0],1]), axis = 1)
print(np.shape(data_counts))
```

```python
user_count, post_count = np.unique(data_counts[:,1],return_counts=True)
print(np.shape(user_count),np.shape(post_count))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(post_count, user_count)
ax.set_xlabel('Number of Posts')
ax.set_ylabel('Number of Users')
ax.set_title('scatter plot')
plt.show()
```
